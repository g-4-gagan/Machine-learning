```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_arr=np.array([1,2,3,4,5])
y_arr=np.array([7,14,15,18,19])
```

```python
def cost_func(y_original, y_predicted):             #calculating cost
    cost = np.sum((y_original - y_predicted)**2)/ (len(y_original))
    return cost

def gradient_descent(x,y,iterations =1000, lr=0.0001, stopping_threshold=1e-6):
    initial_q1 =0.1
    initial_q0 =0.01
    iterations = iterations
    lr=lr
    n=float(len(x))
    prev_cost =None
    
    q1s =[]
    q0s =[]
    for i in range(iterations):
        y_pred =(initial_q1 * x) + initial_q0
        #calucalting the present cost
        present_cost = cost_func(y,y_pred)
        if prev_cost and abs(prev_cost - present_cost) <= stopping_threshold :
            break
            
        prev_cost = present_cost
        q0s.append(present_cost)
        q1s.append(initial_q1)
        
        #partial derivative
        q1_der = -(2/n) * sum(x * (y-y_pred))
        q0_der = -(2/n) * sum(y-y_pred)
        #updating derivative
        initial_q1 = initial_q1 - (lr * q1_der)
        initial_q0 = initial_q0 - (lr * q0_der)
        
    plt.figure(figsize = (10,7))
    plt.plot(q1s,q0s)
    plt.scatter(q1s,q0s,marker='*' ,color='yellow')
    plt.title("Q0 v/s Q1")    
    plt.ylabel("Cost")
    plt.xlabel("Q1")    
    plt.show()
        
    return initial_q1, initial_q0

optimal_q1 ,optimal_q0 = gradient_descent(x_arr,y_arr)
print("Optimal Q1 :: ", optimal_q1)
print("Optimal Q0 :: ", optimal_q0)

Y_pred =optimal_q1 * x_arr + optimal_q0
plt.figure(figsize = (10,7))
plt.scatter(x,y, marker='*' ,color ="green")
plt.plot([min(x_arr), max(x_arr)], [min(Y_pred), max(Y_pred)], color='red',linestyle="dashed")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
    
```
